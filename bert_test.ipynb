```python
from transformers import BertConfig, BertForMaskedLM, BertTokenizerFast, DataCollatorForLanguageModeling, pipeline
```

```python
epochs = 40
# 5000stars_issues_40epochs
# model_path = f"4_proj_issues_{epochs}epochs"
# model_path = "i2000_s1000_40epochs"
# model_path = "gbkwon/issueBERT-large"
model_path = "seBERT"
```

```python
# when the tokenizer is trained and configured, load it as BertTokenizerFast
tokenizer = BertTokenizerFast.from_pretrained(model_path, padding=True)
```

```python
# initialize the model with the config
# model_config = BertConfig(vocab_size=vocab_size, max_position_embeddings=max_length)
# model = BertForMaskedLM.from_pretrained('4_proj_issues_40epochs\checkpoint-180000')
# model = BertForMaskedLM.from_pretrained(model_path + '/test')
model = BertForMaskedLM.from_pretrained(model_path)
```

```python
# when you load from pretrained
#model = BertForMaskedLM.from_pretrained(os.path.join(model_path, "checkpoint-6000"))
#tokenizer = BertTokenizerFast.from_pretrained(model_path)
# or simply use pipeline

fill_mask = pipeline("fill-mask", model=model, tokenizer=tokenizer)
```
